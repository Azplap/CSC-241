```python
%matplotlib inline
import numpy as np
import math, sys
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
plt.rcParams['figure.figsize'] = [15, 8]

#1
#Decide on reasonable physical parameters, e.g. simulate 100 seconds long spike train with 10 spikes per second.

n, p, t = 1, .01, 100000  # number of trials each bin, probability of each trial, tested t times

#2 Should have 10 spikes per second
s = np.random.binomial(n, p, t)
spikes=np.count_nonzero(s)
print("Total # of spikes:", spikes)
plt.title("100 seconds long spike train with 10 spikes per second")
plt.xlabel("Time (ms)")
plt.ylabel("Spike (0 | 1)")
plt.plot(s)
```

### 3) First I found the beta value neccesary to generate 1000 ISI values with the sum of 100.

```python
# generate distance between spikes
# still need 1000 values for 1000 spikes per 100 seconds
# sum of ISIs need to be greater than 100
# need 1000 values x where sum(all x) > 100

#beta = 1/lambda
beta = 0.0001

while(True):
    sum=0
    temp = np.random.exponential(beta, 1000) #scale parameter, size
    sum = np.sum(temp)
    
    if(sum>100):
        print("Correct beta:", beta)
        print("ISI sum:", sum)
        plt.title("Exponentially Generated ISIs")
        plt.plot(temp)     
        break
    
    beta+=0.0001
```

### Plotted values by adding ISIs together one by one and recording a spike for each ISI added.

```python
#3 Continued
# 0.1ms bins still so out of 100,000
exp_dist = np.random.exponential(0.0952, 1000)
spike_times=[]
curr=0
for i in exp_dist:
    curr+=i
    spike_times.append(round(curr, 3))

final=[]
for i in range(0, 100000):
    if i/1000 in spike_times:
        final.append(1)
    else:
        final.append(0)

plt.rcParams['figure.figsize'] = [15, 8]
plt.title("ISI generated 100 seconds long spike train")
plt.xlabel("Time (ms)")
plt.ylabel("Spike (0 | 1)")
plt.plot(final)
print("Spike Count:",np.count_nonzero(final))
        
```

### 4a) Separated values into 100 bins for each seconds.

```python
#4A
#make 100 trials for each second. So 100 trials w/ 1000 values per trial.
count=1
trials=[]
temp=[]
for x in final:
    if(count==1000):
        temp.append(x)
        trials.append(temp)
        count=1
        temp=[]
    else:
        temp.append(x)
        count+=1
```

### 4b) Plotted ISI spikes per trial as red histogram over blue poisson dist. 

```python
#count # of spikes
spikes_per_trial=[]
for trial in trials:
    count=0
    for y in trial:
        if(y==1):
            count+=1
    spikes_per_trial.append(count)

s = np.random.poisson(np.mean(spikes_per_trial), 10000)

plt.title("Blue: Poisson, Red: Spikes Per Trial")
plt.xlabel("Bins")
plt.hist(spikes_per_trial, 14, density=True, color="red")
count, bins, ignored = plt.hist(s, 14, density=True, color="blue")
```

### 4c) Mean was and should be roughly 10, variance was also close to 10. Ratio is close to 1.

```python
#mean and variance
print("Mean:", np.mean(spikes_per_trial))
print("Variance:",np.var(spikes_per_trial))
```

# Computer Problem 2

### a) Created two RFs, one using cos and another using sin

```python
%matplotlib inline
import numpy as np
import math, sys
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt


sigx = 1 #degrees
sigy = 2 #degrees
phi = math.pi / 2
k = 1/0.56

#Gabor Functions
def RF(x,y):
    result = (1/(2 * math.pi * sigx * sigy)) * math.exp(- (x**2)/(2*sigx**2) - (y**2)/(2*sigy**2)) * math.cos(k*x- phi)
    return result

def RFsin(x,y):
    result = (1/(2 * math.pi * sigx * sigy)) * math.exp(- (x**2)/(2*sigx**2) - (y**2)/(2*sigy**2)) * math.sin(k*x- phi)
    return result


# 51x51 Matrix 
increments = [x for x in range(-25, 26, 1)]
myarray = 2* np.asarray(increments)/10


#create gabor matrix for simple cell 1
matrix=[]
for x in myarray:
    temp=[]
    for y in myarray:
        temp.append(RF(y,x))
    matrix.append(temp)

#second RF with sin as well
matrix2=[]
for x in myarray:
    temp=[]
    for y in myarray:
        temp.append(RFsin(y,x))
    matrix2.append(temp)

    
plt.rcParams['figure.figsize'] = [12, 5]    

plt.subplot(1, 2, 1)
plt.title("Gabor Cos")
plt.imshow(matrix)


plt.subplot(1, 2, 2)
plt.title("Gabor Sin")
plt.imshow(matrix2)
```

### X-Axis Cross Section

```python
plt.rcParams['figure.figsize'] = [12, 5]  

plt.subplot(1, 2, 1)
plt.title("X-Axis Cross Section of Gabor Cos")
plt.plot(myarray, matrix[0])


plt.subplot(1, 2, 2)
plt.title("X-Axis Cross Section of Gabor Sin")
plt.plot(myarray, matrix2[0])
```

### b) Create Vertical Grating Image w/ theta=0

```python
#B CREATE IMAGE
sigx = 1 #degrees
sigy = 2 #degrees
phi = math.pi / 2
k = 1/0.56
alpha= math.pi/2
theta=0

def image_formula(x,y, alpha):
    image = math.sin( k*( math.sin(x*math.cos(theta)) + y*math.sin(theta)  ) - alpha )
    return image

# 51x51 Matrix 
increments = [x for x in range(-25, 26, 1)]
myarray = 2* np.asarray(increments)/10

#create vertical grating image I
I=[]
for x in myarray:
    temp=[]
    for y in myarray:
        temp.append(image_formula(y,x, alpha))
    
    I.append(temp)

plt.imshow(I) 
```

### Apply Image to RF

```python
#apply image to RF by multiplication
def combine(m1, I):
    result = []
    for i in range(0, len(m1)):
        #Dot product between RF and Image
        result.append([x * y for x, y in zip(m1[i], I[i])])

    return result
    
    
#total sum of all parts
def mean_spike_rate(x):
    total=0
    for row in x:
        #print(sum(row))
        total+=sum(row)
    return total


result = combine(matrix, I)
result2 = combine(matrix2, I)
```

### Plot mean spike rate as a function of alpha.

```python
#array of alpha values
alpha=[x/100 for x in range(-200, 200)]

def generate_image(alpha):
    I=[]
    for x in myarray:
        temp=[]
        for y in myarray:
            temp.append(image_formula(y,x, alpha))
        I.append(temp)
    return I

#fill mean spike rate list
MSR_list=[]
for x in alpha:
    curr_alpha = x
    image = generate_image(curr_alpha)
    result = combine(matrix, image)
    MSR_list.append((mean_spike_rate(result)**2) * (50/120) )

plt.title("Spike Rate based on varying alpha")
plt.xlabel("Alpha")
plt.ylabel("Mean Spike Rate")
plt.plot(alpha, MSR_list)
```
